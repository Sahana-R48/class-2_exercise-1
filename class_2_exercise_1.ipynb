```python
class Member:
    def __init__(self, name, membershipID):
        self.name = name
        self.membershipID = membershipID

    def display(self):
        print("Name:", self.name)
        print("Membership ID:", self.membershipID)
class RegularMember(Member):
    def __init__(self, name, membershipID):
        super().__init__(name, membershipID)
        self.borrow_limit = 3

    def getBorrowLimit(self):
        return self.borrow_limit

    def display(self):
        print("Member Type: Regular Member")
        print("Name:", self.name)
        print("Borrowing Limit:", self.getBorrowLimit(), "books")
        print()
class PremiumMember(Member):
    def __init__(self, name, membershipID):
        super().__init__(name, membershipID)
        self.borrow_limit = 10

    def getBorrowLimit(self):
        return self.borrow_limit
    def display(self):
        print("Member Type: Premium Member")
        print("Name:", self.name)
        print("Borrowing Limit:", self.getBorrowLimit(), "books")
        print("Additional Feature: Discount on membership fees")
        print()
def main():
    regular = RegularMember("sahana", 101)
    premium = PremiumMember("neevi", 102)

    regular.display()
    premium.display()

if __name__ == "__main__":
    main()
```
