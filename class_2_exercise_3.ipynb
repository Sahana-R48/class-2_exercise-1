```python

class Student:
    def __init__(self, name, roll_number):
        self.name = name
        self.roll_number = roll_number

    def display(self):
        print("\nName:", self.name)
        print("Roll Number:", self.roll_number)

class UndergraduateStudent(Student):
    def __init__(self, name, roll_number, marks):
        super().__init__(name, roll_number)
        self.marks = marks   # list of 3 subject marks

    def calculate_average(self):
        avg = sum(self.marks) / 3
        print("Average Marks (UG):", avg)

class PostgraduateStudent(Student):
    def __init__(self, name, roll_number, marks):
        super().__init__(name, roll_number)
        self.marks = marks   # list of 4 subject marks

    def calculate_average(self):
        avg = sum(self.marks) / 4
        print("Average Marks (PG):", avg)

print("---- Undergraduate Student ----")
name = input("Enter Name: ")
roll = int(input("Enter Roll Number: "))
marks_ug = []

print("Enter marks for 3 subjects:")
for i in range(3):
    marks_ug.append(float(input()))

ug = UndergraduateStudent(name, roll, marks_ug)
ug.display()
ug.calculate_average()


print("\n---- Postgraduate Student ----")
name = input("Enter Name: ")
roll = int(input("Enter Roll Number: "))
marks_pg = []

print("Enter marks for 4 subjects:")
for i in range(4):
    marks_pg.append(float(input()))

pg = PostgraduateStudent(name, roll, marks_pg)
pg.display()
pg.calculate_average()
```
